```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['bank.csv']))
print("✅ Dataset Loaded. Shape:", df.shape)
df.head()
```

```python
# # Drop the target column
# features_df = df.drop(columns=['target'])

# Sample 10% of the data (without target column)
sample_1_percent = df.sample(frac=0.01, random_state=42)

# Save the sampled data to a new CSV file
sample_1_percent.to_csv('sample_10_percent.csv', index=False)

print("10% split saved to 'sample_10_percent.csv' (with target column).")

print(sample_1_percent.shape)
```

```python
features_df = sample_1_percent.drop(columns=['deposit'])
features_df.to_csv('features_df.csv', index=False)
print("10% split saved to 'sample_10_percent.csv' (without target column).")

print(features_df.shape)
```

```python
less_features_df = sample_10_percent.drop(columns=['deposit','contact','housing'])
less_features_df.to_csv('less_features_df.csv', index=False)
print("10% split saved to 'sample_10_percent.csv' (without few columns).")

print(less_features_df.shape)
```
